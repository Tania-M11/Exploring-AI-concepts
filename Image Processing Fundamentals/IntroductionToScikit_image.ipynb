# **Pruebas y experimentación**


```python
import matplotlib.pyplot as plt
from skimage import io, color, filters, morphology
from skimage.util import img_as_ubyte
from skimage.transform import rescale
```

```python
# Load image
image = io.imread('dog.jpeg')

# Prepare the matrix to display the results
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
ax = axes.ravel()

# Show original image
ax[0].imshow(image)
ax[0].set_title('Original')

# Convert image to grayscale
gray_image = color.rgb2gray(image)
ax[1].imshow(gray_image, cmap='gray')
ax[1].set_title('Escala de Grises')

# Apply a Gaussian blur filter
blurred_image = filters.gaussian(gray_image, sigma=1)
ax[2].imshow(blurred_image, cmap='gray')
ax[2].set_title('Desenfoque Gaussiano')

# Apply thresholding
threshold_value = filters.threshold_otsu(gray_image)
binary_image = gray_image > threshold_value
ax[3].imshow(binary_image, cmap='gray')
ax[3].set_title('Umbralización')

# Apply morphological erosion
eroded_image = morphology.erosion(binary_image, morphology.square(3))
ax[4].imshow(eroded_image, cmap='gray')
ax[4].set_title('Erosión Morfológica')

# Apply morphological dilation
dilated_image = morphology.dilation(binary_image, morphology.square(3))
ax[5].imshow(dilated_image, cmap='gray')
ax[5].set_title('Dilatación Morfológica')

# Resize the image
rescaled_image = rescale(image, scale=0.5, multichannel=True)
ax[6].imshow(rescaled_image)
ax[6].set_title('Imagen Redimensionada')

# Show borders
edges = filters.sobel(gray_image)
ax[7].imshow(edges, cmap='gray')
ax[7].set_title('Bordes')

# Show the image inverted
inverted_image = img_as_ubyte(1 - gray_image)
ax[8].imshow(inverted_image, cmap='gray')
ax[8].set_title('Imagen Invertida')

# Adjust the layout of the figure
for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
