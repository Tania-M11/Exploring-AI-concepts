#Implementación una Red Neuronal Convolucional (CNN) aplicada a los conjuntos de datos CIFAR10 y CIFAR100 usando Keras

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10, cifar100
```

```python
# Cargar el conjunto de datos CIFAR-10
(train_images_10, train_labels_10), (test_images_10, test_labels_10) = cifar10.load_data()

# Cargar el conjunto de datos CIFAR-100
(train_images_100, train_labels_100), (test_images_100, test_labels_100) = cifar100.load_data()
```

```python
# Normalizar los valores de píxeles de las imágenes para CIFAR-10
train_images_10 = train_images_10.astype('float32') / 255.0
test_images_10 = test_images_10.astype('float32') / 255.0

# Normalizar los valores de píxeles de las imágenes para CIFAR-100
train_images_100 = train_images_100.astype('float32') / 255.0
test_images_100 = test_images_100.astype('float32') / 255.0
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir el modelo CNN para CIFAR-10
model_cifar10 = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compilar el modelo para CIFAR-10
model_cifar10.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# Mostrar un resumen del modelo para CIFAR-10
model_cifar10.summary()
```

#Construcción del Modelo de Clasificación


```python
# Modelo para CIFAR-10
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir el modelo CNN para CIFAR-10
model_cifar10 = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compilar el modelo para CIFAR-10
model_cifar10.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# Mostrar un resumen del modelo para CIFAR-10
model_cifar10.summary()
```

```python
# para CIFAR-100
model_cifar100 = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(100, activation='softmax')  # 100 clases para CIFAR-100
])

# Compilar el modelo para CIFAR-100
model_cifar100.compile(optimizer='adam',
                        loss='sparse_categorical_crossentropy',
                        metrics=['accuracy'])

# Mostrar un resumen del modelo para CIFAR-100
model_cifar100.summary()
```

```python
#Compilar modelos

# Para CIFAR-10
model_cifar10.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])


# Para CIFAR-100
model_cifar100.compile(optimizer='adam',
                        loss='sparse_categorical_crossentropy',
                        metrics=['accuracy'])
```

#Entrenar modelos

```python
 #para CIFAR-10
history_cifar10 = model_cifar10.fit(train_images_10, train_labels_10, epochs=10, validation_split=0.2)
```

```python
# Para CIFAR-100
history_cifar100 = model_cifar100.fit(train_images_100, train_labels_100, epochs=10, validation_split=0.2)
```

#Evaluación del Modelo

```python
# Para CIFAR-10
train_loss_cifar10, train_accuracy_cifar10 = model_cifar10.evaluate(train_images_10, train_labels_10, verbose=2)
test_loss_cifar10, test_accuracy_cifar10 = model_cifar10.evaluate(test_images_10, test_labels_10, verbose=2)

print(f'Pérdida en conjunto de entrenamiento (CIFAR-10): {train_loss_cifar10:.4f}')
print(f'Precisión en conjunto de entrenamiento (CIFAR-10): {train_accuracy_cifar10:.4f}')
print(f'Pérdida en conjunto de prueba (CIFAR-10): {test_loss_cifar10:.4f}')
print(f'Precisión en conjunto de prueba (CIFAR-10): {test_accuracy_cifar10:.4f}')
```

```python
# Para CIFAR-100
train_loss_100, train_accuracy_100 = model_cifar100.evaluate(train_images_100, train_labels_100, verbose=2)
test_loss_100, test_accuracy_100 = model_cifar100.evaluate(test_images_100, test_labels_100, verbose=2)

print(f'Pérdida en conjunto de entrenamiento (CIFAR-100): {train_loss_100:.4f}')
print(f'Precisión en conjunto de entrenamiento (CIFAR-100): {train_accuracy_100:.4f}')
print(f'Pérdida en conjunto de prueba (CIFAR-100): {test_loss_100:.4f}')
print(f'Precisión en conjunto de prueba (CIFAR-100): {test_accuracy_100:.4f}')

```

#Visualización de Resultados

```python
def plot_training_results(history, dataset_name):
    plt.figure(figsize=(12, 4))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Training and Validation Accuracy ({dataset_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Training and Validation Loss ({dataset_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Visualizar resultados del entrenamiento para CIFAR-10
plot_training_results(history_cifar10, 'CIFAR-10')

# Visualizar resultados del entrenamiento para CIFAR-100
plot_training_results(history_cifar100, 'CIFAR-100')
```

#Ejemplos de Imágenes de Prueba con Predicciones

```python

# Para CIFAR-10
def show_prediction_examples_cifar10(images, labels, model, class_names, num_examples=5):
    plt.figure(figsize=(15, 3*num_examples))
    for i in range(num_examples):
        plt.subplot(num_examples, 2, 2*i + 1)
        plt.imshow(images[i])
        plt.axis('off')
        true_label = class_names[int(labels[i])]
        plt.title(f'True Label: {true_label}')

        plt.subplot(num_examples, 2, 2*i + 2)
        img = np.expand_dims(images[i], axis=0)
        prediction = model.predict(img)
        predicted_label = class_names[np.argmax(prediction)]
        confidence = np.max(prediction)
        plt.bar(range(len(class_names)), prediction.ravel())
        plt.xticks(range(len(class_names)), class_names, rotation=45)
        plt.title(f'Predicted Label: {predicted_label}\nConfidence: {confidence:.2f}')

    plt.tight_layout()
    plt.show()

# Mostrar ejemplos de imágenes de prueba con predicciones para CIFAR-10
show_prediction_examples_cifar10(test_images_10, test_labels_10, model_cifar10, cifar10_labels)
```

```python
# Para CIFAR-100
def show_example_prediction_cifar100(image, label, model, class_names):
    plt.figure(figsize=(30, 26))

    # Imagen de prueba
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.axis('off')
    plt.title(f'Imagen de prueba (Etiqueta: {class_names[label]})')

    # Predicción del modelo
    img = np.expand_dims(image, axis=0)
    prediction = model.predict(img)
    predicted_label = np.argmax(prediction)

    # Distribución de las probabilidades de predicción
    plt.subplot(1, 2, 2)
    y_pos = np.arange(len(class_names)) * 7.0
    plt.barh(y_pos, prediction.ravel(), align='center')
    plt.yticks(y_pos, class_names, fontsize=20)
    plt.xlabel('Probabilidad')
    plt.title(f'Predicción: {class_names[predicted_label]}')
    plt.gca().invert_yaxis()
    plt.tight_layout()

    plt.show()


# Elegir una imagen de prueba y mostrar la predicción para CIFAR-100

index = 123  # Probar con diferentes imágenes
image = test_images_100[index]
true_label = int(test_labels_100[index])


# Mostrar la imagen de prueba y la predicción
show_example_prediction_cifar100(image, true_label, model_cifar100, cifar100_labels)
```
