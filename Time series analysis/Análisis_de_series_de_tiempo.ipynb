```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#Generar datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start = '2022-01-01', end='2022-12-31')
values = np.arange(len(dates))*.2 + np.random.rand(len(dates))*50
ts = pd.Series(values, index=dates)
```

```python
#Graficar la serie original
plt.figure(figsize=(10,5))
plt.plot(ts, label = 'Serie Original')
plt.legend()
plt.show()
```

```python
#Descomposicin de la serie de tiempo
descomposition = seasonal_decompose(ts, model='additive')
```

```python
#Grfica de la serie original
plt.figure(figsize=(10,5))
plt.plot(ts, label = 'Serie Original')
plt.legend()
plt.show()

#Grafica de la tendencia
plt.subplot(412)
plt.plot(descomposition.trend, label = 'Tendencia')
plt.legend()

#Grfica de la estacionaidad
plt.subplot(413)
plt.plot(descomposition.seasonal, label = 'Estacionaridad')
plt.legend()

#Grfica de los residuos
plt.subplot(414)
plt.plot(descomposition.resid, label = 'Residuos')
plt.legend()

plt.tight_layout()
plt.show()
```
